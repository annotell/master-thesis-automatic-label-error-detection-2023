```python
def plot_issues(figpath, main_title, data_imgs, labels_gt, id_iter, nrows, ncols, labels_pred, losses, imgnames, labels_gt_des, labels_pred_des, probs, show_img=True):
#     print(f'id_iter={id_iter}')
#     plt.figure()
    dpi = 400
    fig = plt.figure(dpi=dpi, figsize=(20,18))
    
    for count, id in enumerate(id_iter):
        # print(f'id={id}, count={count}')
        # print(f'imgnames={imgnames[id]}')
        # # print(f'data_imgs={data_imgs[id]}')
        # print(f'labels_gt={labels_gt[id]}')
        # print(f'labels_gt_des={labels_gt_des[id]}')
        # print(f'probs={probs[id]}')
        
        title = f"{imgnames[id]}\ngt({labels_gt[id]}): {labels_gt_des[id]}\nprob: {probs[id]:.3f}"
        ax = fig.add_subplot(nrows, ncols, count + 1)
        ax.set_title(title, fontdict={'fontsize': 8, 'fontweight': 4})
        plt.imshow(data_imgs[id], cmap="gray")
        #         plt.title(title, fontdict={'fontsize': 4, 'fontweight': 4})
        plt.axis("off")
        # plt.subplots_adjust(wspace=0.4, hspace=0.4, left=0.1, bottom=1, right=0.95, top=3.5)
#         print('----')
    plt.tight_layout()
    # print(f'# of pyplot figure stack: {plt.get_fignums()}')
    # plt.suptitle(main_title)
    plt.savefig(figpath, dpi=dpi, bbox_inches='tight')

    print(f'show_img={show_img} {type(show_img)}')
    if show_img:
        plt.show()
    # print(f'# of pyplot figure stack: {plt.get_fignums()}')
    
```

```python
# https://github.com/pytorch/pytorch/issues/16797#issuecomment-633423219
class CPU_Unpickler(pickle.Unpickler):
    def find_class(self, module, name):
        if module == 'torch.storage' and name == '_load_from_bytes':
            return lambda b: torch.load(io.BytesIO(b), map_location='cpu')
        else: return super().find_class(module, name)
```

```python

def cls_to_probs(saved_obj):
    cls_score = saved_obj['cls_scores']
    # print(f'[cls_to_probs] cls_score.shape={cls_score.shape}')

    probs = F.softmax(cls_score, dim=-1).cpu().numpy()
    # print('cls_score.shape=', cls_score.shape)
    # print(f'scores.shape={probs.shape}')
    assert cls_score.size(0) == saved_obj['gt_bboxes'].size(0)

    # sanity check
    # close_to_1 = probs.sum(axis=1)
    # print(f'close_to_1={close_to_1}')

    return probs

def crop_bbox(xmin, ymin, xmax, ymax, img, img_shape, pad=0):
    img_xmax = img_shape[0]
    img_ymax = img_shape[1]
    xmin = int(xmin) - pad if int(xmin) - pad > 0 else 0
    ymin = int(ymin) - pad if int(ymin) - pad > 0 else 0
    xmax = int(xmax) + pad if int(xmin) + pad > 0 else img_xmax
    ymax = int(ymax) + pad if int(ymax) + pad > 0 else img_ymax

    # print(f'img.shape={img.shape}')
    cropped = img[xmin:xmax, ymin:ymax, :]
    # print(f'cropped.shape={cropped.shape}')
    return cropped


def crop_feature(xmin, ymin, xmax, ymax, feat, feat_shape, img_shape, pad_percent):
    feat_xmax = feat_shape[0]
    feat_ymax = feat_shape[1]
    xmin = (xmin / img_shape[0]) * feat_shape[0]
    xmax = (xmax / img_shape[0]) * feat_shape[0]
    ymin = (ymin / img_shape[1]) * feat_shape[1]
    ymax = (ymax / img_shape[1]) * feat_shape[1]
    x_pad = int(pad_percent * (xmax - xmin))
    y_pad = int(pad_percent * (ymax - ymin))

    xmin = int(xmin) - x_pad if int(xmin) - x_pad > 0 else 0
    ymin = int(ymin) - y_pad if int(ymin) - y_pad > 0 else 0
    xmax = int(xmax) + x_pad if int(xmin) + x_pad > 0 else feat_xmax
    ymax = int(ymax) + y_pad if int(ymax) + y_pad > 0 else feat_ymax

    # print(f'img.shape={img.shape}')
    # print(f'[crop_feature] feat.shape={feat.shape}')
    # print(f'[crop_feature] x range=({xmin}, {xmax}), y range=({ymin}, {ymax})')
    cropped = feat[:, xmin:xmax, ymin:ymax]
    # print(f'cropped.shape={cropped.shape}')
    return cropped


def draw_bboxes(img, bboxes, labels, labels_des, name2RGB):

    for idx, (bbox, label, label_des) in enumerate(zip(gt_bboxes, gt_labels, labels_des)):
        ymin, xmin, ymax, xmax = [int(i) for i in bbox]
        
        # start = (xmin, ymin)
        # end = (xmax, ymax)
        r = [xmin,xmax,xmax,xmin, xmin]
        c = [ymax,ymax,ymin,ymin, ymax]
        rr, cc = polygon_perimeter(r, c, img.shape)
        # rr, cc = polygon_perimeter(start, end=end, shape=img.shape)
        rgb = name2RGB[label_des]
        img[rr, cc ,0] = rgb[0]
        img[rr, cc ,1] = rgb[1]
        img[rr, cc ,2] = rgb[2]

        font = cv2.FONT_HERSHEY_PLAIN
        text_org = (ymin, xmin-2)

        if label_des == 'traffic sign':
            label_des = 'sign'
        elif label_des == 'traffic light':
            label_des = 'light'
        cv2.putText(img, label_des, text_org, font, 0.5, rgb, 1)

    # print(f'img.shape={img.shape}')
    # img = np.moveaxis(img, 0, -1)
    
    # dpi = 150
    # fig = plt.figure(dpi=dpi, figsize=(12,14))
    # plt.imshow(img)


def crop_bboxes(img, img_name, gt_bboxes, gt_labels, pad):
    bbox_imgs = []

    for idx, (bbox, labels) in enumerate(zip(gt_bboxes, gt_labels)):
        ymin, xmin, ymax, xmax = bbox
        
        bbox_img = crop_bbox(xmin, ymin, xmax, ymax, img, img_shape, pad)
        bbox_imgs.append(bbox_img)

    return bbox_imgs

def crop_features(feature, img_name, img_shape, gt_bboxes, gt_labels, pad_percent):
    bbox_feat0s = []
    bbox_feat1s = []
    bbox_feat2s = []
    bbox_feat3s = []
    bbox_feat4s = []
    # print(f'len of input feature={len(feature)}')

    feat0s = np.squeeze(feature[0], axis=0).detach().cpu().numpy()
    feat1s = np.squeeze(feature[1], axis=0).detach().cpu().numpy()
    feat2s = np.squeeze(feature[2], axis=0).detach().cpu().numpy()
    feat3s = np.squeeze(feature[3], axis=0).detach().cpu().numpy()
    feat4s = np.squeeze(feature[4], axis=0).detach().cpu().numpy()

    # print(f'shape of feat0s={feat0s.shape}')
    # print(f'shape of feat1s={feat1s.shape}')
    # print(f'shape of feat2s={feat2s.shape}')
    # print(f'shape of feat3s={feat3s.shape}')
    # print(f'shape of feat4s={feat4s.shape}')
    feat0_shape = feat0s[0].shape
    feat1_shape = feat1s[0].shape
    feat2_shape = feat2s[0].shape
    feat3_shape = feat3s[0].shape
    feat4_shape = feat4s[0].shape


    for idx, (bbox, labels) in enumerate(zip(gt_bboxes, gt_labels)):
        ymin, xmin, ymax, xmax = bbox
        
        bbox_feat0 = crop_feature(xmin, ymin, xmax, ymax, feat0s, feat0_shape, img_shape, pad_percent)
        bbox_feat1 = crop_feature(xmin, ymin, xmax, ymax, feat1s, feat1_shape, img_shape, pad_percent)
        bbox_feat2 = crop_feature(xmin, ymin, xmax, ymax, feat2s, feat2_shape, img_shape, pad_percent)
        bbox_feat3 = crop_feature(xmin, ymin, xmax, ymax, feat3s, feat3_shape, img_shape, pad_percent)
        bbox_feat4 = crop_feature(xmin, ymin, xmax, ymax, feat4s, feat4_shape, img_shape, pad_percent)
        bbox_feat0s.append(bbox_feat0)
        bbox_feat1s.append(bbox_feat1)
        bbox_feat2s.append(bbox_feat2)
        bbox_feat3s.append(bbox_feat3)
        bbox_feat4s.append(bbox_feat4)

    print([i.shape for i in bbox_feat0s])
    bbox_feat0s = np.array(bbox_feat0s)
    bbox_feat1s = np.array(bbox_feat1s)
    bbox_feat2s = np.array(bbox_feat2s)
    bbox_feat3s = np.array(bbox_feat3s)
    bbox_feat4s = np.array(bbox_feat4s)

    return bbox_feat0s, bbox_feat1s, bbox_feat2s, bbox_feat3s, bbox_feat4s


def show_feature(feats):
    # if feats[0]
    feat0 = np.squeeze(feats[0], axis=0).detach().cpu().numpy()
    feat1 = np.squeeze(feats[1], axis=0).detach().cpu().numpy()

    limit = 3
    for i in range(len(feat0)):
        fig = plt.figure(dpi=60, figsize=(12,14))
        plt.imshow(feat0[i])
        plt.show()
        if i == limit:
            break
    print('------')
    for i in range(len(feat1)):
        fig = plt.figure(dpi=60, figsize=(12,14))
        plt.imshow(feat1[i])
        plt.show()
        if i == limit:
            break



img_features = []
bbox_probs = []
bbox_max_probs = []
bbox_imgs = []
bbox_imgnames = []
bbox_imgpaths = []
bbox_pred_labels = []
bbox_pred_labels_des = []
bbox_gt_labels = []
bbox_gt_labels_des = []

bbox_feats_all = None
# bbox_per_feat0s = np.empty_like(saved_objs[0]['feature'][0].detach().cpu().numpy().squeeze(0))
# bbox_per_feat1s = np.empty_like(saved_objs[0]['feature'][1].detach().cpu().numpy().squeeze(0))
# bbox_per_feat2s = np.empty_like(saved_objs[0]['feature'][2].detach().cpu().numpy().squeeze(0))
# bbox_per_feat3s = np.empty_like(saved_objs[0]['feature'][3].detach().cpu().numpy().squeeze(0))
# bbox_per_feat4s = np.empty_like(saved_objs[0]['feature'][4].detach().cpu().numpy().squeeze(0))
# print(f'0 shape of bbox_per_feat0s = {bbox_per_feat0s.shape}')
# print(f'0 shape of bbox_per_feat1s = {bbox_per_feat1s.shape}')
# print(f'0 shape of bbox_per_feat2s = {bbox_per_feat2s.shape}')
# print(f'0 shape of bbox_per_feat3s = {bbox_per_feat3s.shape}')
# print(f'0 shape of bbox_per_feat4s = {bbox_per_feat4s.shape}')





for idx, saved_obj in enumerate(saved_objs):

    # if idx != 0:
    #     break
    
    # print(list(saved_obj.keys()))
    # print('gt_bboxes.shape=', saved_obj['gt_bboxes'].shape)
    # print(f'0 shape of bbox_feats = {saved_obj["bbox_feats"].shape}')
    # break

    # print(saved_obj)(244, 67, 54)
    img_path = saved_obj['img_path']
    img_ori = np.array(Image.open(img_path))
    img_shape = saved_obj['img_shape']
    img_name = saved_obj['img_name']
    # img_feature = saved_obj['feature']
    gt_bboxes = saved_obj['gt_bboxes'].detach().cpu().tolist()
    gt_labels = saved_obj['gt_labels'].detach().cpu().tolist()
    bbox_feats = saved_obj['bbox_feats'].detach().cpu()

    probs = cls_to_probs(saved_obj)
    # print(f'probs.shape={probs.shape}')
    maxprob_idx = np.argmax(probs, axis=1)
    maxprob = np.max(probs, axis=1)
    bbox_pred_label = maxprob_idx # index is the label number


    assert probs.shape[0] == maxprob.shape[0]

    gt_labels_des = [id2class[i] for i in gt_labels]
    bbox_pred_label_des = [id2class[i] for i in bbox_pred_label]
    # print(f'bbox_pred_labels_des={bbox_pred_labels_des}')
    # break

    # print(f'len of img_feature={len(img_feature)}')
    # print(f'img_feature[0] shape= {img_feature[0].shape}')
    # print(f'img_feature[1] shape= {img_feature[1].shape}')
    # show_feature(img_feature)

    img_with_bboxes = draw_bboxes(img_ori, gt_bboxes, gt_labels, gt_labels_des, name2RGB)

    bbox_per_imgs = crop_bboxes(img_ori, img_name, gt_bboxes, gt_labels, pad=20)
    # bbox_per_feat0, bbox_per_feat1, bbox_per_feat2, bbox_per_feat3, bbox_per_feat4 = crop_features(img_feature, img_name, img_shape, gt_bboxes, gt_labels, pad_percent=0)
    
    gt_labels_des = [id2class[i] for i in gt_labels]
    # print(f'number of bbox: {len(bbox_per_imgs)}')
    

    bbox_probs.extend(probs)
    bbox_pred_labels.extend(bbox_pred_label)
    bbox_pred_labels_des.extend(bbox_pred_label_des)
    bbox_max_probs.extend(maxprob)
    bbox_imgs.extend(bbox_per_imgs)
    bbox_imgnames.extend([img_name]*len(bbox_per_imgs))
    bbox_imgpaths.extend([img_path]*len(bbox_per_imgs))
    bbox_gt_labels.extend(gt_labels)
    bbox_gt_labels_des.extend(gt_labels_des)
    # bbox_per_feat0s = np.concatenate((bbox_per_feat0s, bbox_per_feat0), axis=0)
    # bbox_per_feat1s = np.concatenate((bbox_per_feat1s, bbox_per_feat1), axis=0)
    # bbox_per_feat2s = np.concatenate((bbox_per_feat2s, bbox_per_feat2), axis=0)
    # bbox_per_feat3s = np.concatenate((bbox_per_feat3s, bbox_per_feat3), axis=0)
    # bbox_per_feat4s = np.concatenate((bbox_per_feat4s, bbox_per_feat4), axis=0)

    if bbox_feats_all is None:
        bbox_feats_all = bbox_feats
        continue
    bbox_feats_all = np.concatenate((bbox_feats_all, bbox_feats), axis=0)


# losses=[]
# df = pd.DataFrame(zip(bbox_max_probs, bbox_gt_labels, bbox_gt_labels_des, bbox_pred_labels, bbox_pred_labels_des, losses, bbox_imgpaths, bbox_imgnames),
#                   columns=['probs', 'labels_gt', 'labels_gt_des', 'labels_pred', 'labels_pred_des', 'losses', 'paths', 'imgnames'])
# print(df.head(10))

bbox_probs = np.array(bbox_probs)
bbox_max_probs = np.array(bbox_max_probs)
bbox_imgs = np.array(bbox_imgs)
bbox_gt_labels = np.array(bbox_gt_labels)
bbox_gt_labels_des = np.array(bbox_gt_labels_des)
bbox_imgnames = np.array(bbox_imgnames)

# print(f'len of bbox_per_feat0s={len(bbox_per_feat0s)}')
# print(f'len of bbox_per_feat1s={len(bbox_per_feat1s)}')
# print(f'len of bbox_per_feat2s={len(bbox_per_feat2s)}')
# print(f'len of bbox_per_feat3s={len(bbox_per_feat3s)}')
# print(f'len of bbox_per_feat4s={len(bbox_per_feat4s)}')
# print(f'shape of bbox_per_feat0s[0] = {bbox_per_feat0s[0].shape}')
# print(f'shape of bbox_per_feat1s[0] = {bbox_per_feat1s[0].shape}')
# print(f'shape of bbox_per_feat2s[0] = {bbox_per_feat2s[0].shape}')
# print(f'shape of bbox_per_feat3s[0] = {bbox_per_feat3s[0].shape}')
# print(f'shape of bbox_per_feat4s[0] = {bbox_per_feat4s[0].shape}')

# bbox_per_feat0s = np.vstack(bbox_per_feat0s)
# bbox_per_feat1s = np.vstack(bbox_per_feat1s)
# bbox_per_feat2s = np.vstack(bbox_per_feat2s)
# bbox_per_feat3s = np.vstack(bbox_per_feat3s)
# bbox_per_feat4s = np.vstack(bbox_per_feat4s)

# print(f'shape of bbox_probs = {bbox_probs.shape}')
# print(f'shape of bbox_max_probs = {bbox_max_probs.shape}')
# print(f'shape of bbox_imgs = {bbox_imgs.shape}')
# print(f'shape of bbox_gt_labels = {bbox_gt_labels.shape}')
# print(f'shape of bbox_gt_labels_des = {bbox_gt_labels_des.shape}')
# print(f'shape of bbox_imgnames = {bbox_imgnames.shape}')

# print(f'shape of bbox_per_feat0s = {bbox_per_feat0s[0].shape}')
# print(f'shape of bbox_per_feat1s = {bbox_per_feat1s[0].shape}')
# print(f'shape of bbox_per_feat2s = {bbox_per_feat2s[0].shape}')
# print(f'shape of bbox_per_feat3s = {bbox_per_feat3s[0].shape}')
# print(f'shape of bbox_per_feat4s = {bbox_per_feat4s[0].shape}')
print(f'shape of bbox_feats_all = {bbox_feats_all.shape}')
```

```python
# https://github.com/rom1504/image_embeddings/blob/e6233e78b857285c32755587f585fa251cabae17/image_embeddings/knn/knn.py
def build_index(emb_flatten, dist):
    #print(f'emb.shape={emb_flatten.shape}')
    d = emb_flatten.shape[1]
    xb = emb_flatten

    if dist == 'inner product':
        # cosine similarity
        index = faiss.IndexFlatIP(d)
    elif dist == 'euclidean':
        index = faiss.IndexFlatL2(d)

    index.add(xb)
    return index


def search(index, id_to_name, emb, k):
    D, I = index.search(np.expand_dims(emb, 0), k)  # actual search
    return list(zip(D[0], [id_to_name[x] for x in I[0]]))

# embs = bbox_feats_all
embs_flatten = np.reshape(bbox_feats_all, (bbox_feats_all.shape[0], bbox_feats_all.shape[1]*bbox_feats_all.shape[2]*bbox_feats_all.shape[3]))
embs_256 = bbox_feats_all[:,:, 0, 0]
embs_7x7 = bbox_feats_all[:,0, :, :]
embs_7x7 = np.reshape(embs_7x7, (embs_7x7.shape[0], embs_7x7.shape[1]*embs_7x7.shape[2]))
embs = embs_flatten

# before normalization
index_ip = build_index(embs, dist='inner product')

embs = normalize(embs, norm="l2", axis=1)
print(f'embs.shape={embs.shape}')
print(f'embs should be unit vector = {np.sum(embs**2, axis=1)}')

index_l2 = build_index(embs, dist='euclidean')
index_cos = build_index(embs, dist='inner product')

```
